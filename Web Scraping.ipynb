```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_horror_films"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
soup
```

```python
soup.prettify
```

```python
table = soup.find('table')
countries = []
for row in table.find_all('tr')[2:]:
    temp = row.text.replace('\n\n', " ").strip()
    temp_list = temp.split()
    countries.append((temp_list))
countries
```

```python
new_table=[]
for row in table.find_all('tr')[1:-1]:
    columns = row.find_all('td')
    new_table.append([column.text for column in columns[0:]])
new_table
```

# Task 1

```python
import pandas as pd
```

```python
df = pd.DataFrame(new_table, columns=['Rank','Film', 'worldwidegross', 'year','Franchise or basis','ref'])
df['ref'] = df['ref'].str.replace('\n','')
df
```

```python
data = df.drop(labels=range(20,49),axis=0)
data
```

```python
data['worldwidegross'] = data['worldwidegross'].str.replace('$','').astype(str)
```

```python
data['worldwidegross'] = data['worldwidegross'].str.replace(',', '')
data['worldwidegross'] = data['worldwidegross'].astype(int)
```
